# Graph editing of Detroit 
### Adding attributes to the road network ready for the ABM

In this basic ABM implementation, police vehicles move along the road retwork to respond to calls for service. They head back to the station when the incident is sorted.

The movement of the agent is basic: they move from node to node along a route (shortest path to thyeir target). This can be improved later by making them move along the geometry of the graph.

Required libraries:
- OSMNX
- MESA (although can be done without, just need a little tweak)
- Imageio for creating GIFs (although can be done differently, I just haven't worked on it yet)

Model files:
- AgentFramework2: contains class object Agent
- IncidentFramework: contains class object Incident
- ModelFrameword: contains class object Model (main model)

There are 2 version of the same graph of Detroit (multidigraph):
    - G: is a latlong graph (with gdf_nodes and gdf_edges)
    - G_proj is a UTM (metric) (with gdf_nodes_proj and gdf_edges_proj)

### Convert maxspeed to kph

```python
# Replace NaN with 25 mph default
gdf_edges.loc[gdf_edges['maxspeed'].isnull(),'maxspeed'] = '25 mph'
gdf_edges['maxspeed'].head()
```

```python
def mph2kmph(mph):
    if type(mph) is list:
        # select first element of the list
        mph = mph[0]
        # extract speed numerical value 
        mph = float(mph.split(" ", 1)[0])
        # convert to kph
        kph = mph * 1.60934
    else:
        # extract speed numerical value 
        mph = float(mph.split(" ", 1)[0])
        # convert to kph
        kph = mph * 1.60934
    return kph
```

```python
%%time
gdf_edges['maxspeed_kph'] = [mph2kmph(mph) for index, mph in gdf_edges['maxspeed'].iteritems()]
```

### Calculate edge driving time (in minutes)

```python
def calculateTime(row) :
    
    # Convert kph to meter per minute
    meters_per_minute = row['maxspeed_kph'] * 1000 / 60 #km per hour to m per minute

    # add an edge attribute for time in minutes required to traverse each edge
    return row['length'] / meters_per_minute
```

```python
%%time
gdf_edges['drive_time'] = [calculateTime(edge) for index, edge in gdf_edges.iterrows()]
gdf_edges['drive_time'].head()
```

#### Function to plot street network and specific edges

```python
def plot_edges(gdf_edges, streets_within_buffer=None, current_location_node = None):
    """ Function that plots the geometry LINES of particular edges (route, hot edges etc...)"""
    if streets_within_buffer:
        liste_pairs = streets_within_buffer.index.tolist()
        colors = ['red' if element[0] in liste_pairs else 'grey' for element in gdf_edges.iterrows()]
        widths = [2 if element[0] in liste_pairs else 0.5 for element in gdf_edges.iterrows()]
        f, ax = plt.subplots(1, figsize=(20, 20))
        gdf_edges.plot(alpha=1, axes=ax, linewidth=widths, edgecolor=colors)
    else:
        # just plotting road network, without hot edges
        f, ax = plt.subplots(1, figsize=(20, 20))
        gdf_edges.plot(alpha=1, axes=ax, colors = ['grey' for element in gdf_edges.iterrows()])
    
    # plot the point
    if current_location_node:
        print(current_location_node)
        current_location = gdf_nodes.loc[current_location_node].geometry.coords[0]
        print(current_location)
        ax.scatter(current_location[0],current_location[1], c='green', s = 100)
    plt.show()
```
