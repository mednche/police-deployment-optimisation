# Data pre-processing

The model is designed to be extensible to any urban environment and allows the user to choose a specific locality. The road network for this chosen locality is downloaded from OpenStreetMap and automatically converted to a simplified NetworkX graph using the Python package [OSMNX](https://osmnx.readthedocs.io/en/stable/). 

In the NetworkX graph, nodes correspond to intersections and edges represent segments of uninterrupted road between two intersections. When downloading a road network in OSMNX, specifying the option network type = ‘drive’ creates a directed graph in which edges can only be used in a certain direction. In such a graph, one-way roads can be represented as a single unidirectional edge, while two-way roads require two unidirectional edges (one for each direction).

OpenStreetMap was chosen in this thesis as it is an open source dataset in comparison to other alternatives (e.g. the Integrated Transport Network). Although the crowd sourcing nature of the dataset means it is harder to control for a consistent quality, its free accessibility brings significant reproducible benefits in a research environment, especially when combined with the convenience of the OSMNX package.

**STEP 1: Preprocess the road network**

1. Download and save the road network using OSMNX

2. Add extra columns to the geodataframes of nodes and edges

Edges:
- calculate `travel_time` and `travel_time_mins` (precalculted for faster calculations in ABM)
- find the patrol beat that each street falls into
    
Nodes:
- calculate `total_length_adj_edges` (used when getting hot streets to patrol based on historical incidents)

3. Convert the spatial coordinates for the location of your incidents and crimes to graph nodes.

**STEP 2: Preprocess the patrol beats**

Once the road network graph has been saved, we need to add a `centroid_node` column to the `patrol_beats `

**STEP3: Preprocess incidents and crimes datasets**

Similarily, we need to add a `node` column to the `incidents` and `crimes` datasets. Because of the size of these datasets, we need to run this last step using multiprocessing.

## Edit nodes and edges geodataframes

```python
## Automatic edit of the gdfs using OSMNX

# Add edge speeds (km per hour) to graph as new speed_kph edge attributes.
G = ox.speed.add_edge_speeds(G)
# Add edge travel time (seconds) to graph as new travel_time edge attributes.
G = ox.speed.add_edge_travel_times(G)
```

In a typical graph, nodes are not spatially bounded, but defined in relation to their neighbouring nodes. As a result, the graph does not contain information concerning the spatial location of the real world intersection that are represented by the graph nodes. In order for the graph to be embedded into 2D space, OSMNX couples the graph with two datasets: one for its nodes and one for its edges.

```python
# For manual edits, we need to get the gdfs from graph
gdf_nodes, gdf_edges = ox.graph_to_gdfs(G)
```

```python
gdf_nodes.head()
```

The edges dataset is composed of information extracted from the OSMNX edge dataset (`osmid`, `geometry`, etc.).

`geometry` represents the spatial location of the corresponding road intersections in the form of a pair of spatial coordinates. This attribute is used in the model to identify the closest graph node to each occurring CFS incident.

`osmid` represents the unique OSM id for this node.

```python
gdf_edges.head()
```

The edges dataset is composed of information extracted from the OSMNX edge dataset (`osmid`, `u`, `v`, `geometry`, `oneway`, `length`, `speed_kph`, etc.).

G is a directed graph. In this type of graph, an edge runs from a start node (called `u` in OMSNX) and an end node (called `v` in OSMNX).

The `geometry` attribute of each edge allows one to faithfully recreate the true spatial shape and location of the road segment represented by each edge. This attribute takes the form of a linestring object: a list of at least 2 points (which are themselves pairs of latitude and longitude coordinates). This attribute is used in the model to identify the patrol beat that each edge of the road network belongs to.

### 1. Edges

Let's now add the attributes `travel_time_mins` and `patrol_beat` to the edge dataset.

#### Create a travel_time_mins

```python
gdf_edges['travel_time_mins'] = gdf_edges['travel_time']/60
```

#### Find the patrol beat that each edge fall into based on geometry

```python
from shapely.geometry import Point
```

Note: this may take up to a minute

```python
def get_beat_for_edge(edge, patrol_beats) :
    """ This function identify the patrol beat on which a given edge is falling.
    It uses shapely's intersects function"""
    
    # Most of the time, just use the centroid
    candidate_beats = patrol_beats[patrol_beats.contains(edge.geometry.centroid)]['name']
    
    if len(candidate_beats) == 0:
        # Street centroid is outside of all Detroit patrol beats boundaries
        # for those only, need to make the buffer the boundaries!
        beat_num= None
        for index, beat in patrol_beats.iterrows():
            poly = beat.geometry.buffer(+1e-3)
            if poly.intersects(edge.geometry):
                beat_num= beat['name']
                break
            
    elif len(candidate_beats) > 1:
        print('MORE THAN ONE BEAT FOR CENTROID {}!'.format(edge))
        #break
    
    else:
        # Normal case: street centroid falls in only one patrol beat 
        beat_num= candidate_beats.iloc[0]
            
    
    return beat_num

patrol_beats = gpd.read_file('../data/patrol_beats/patrol_beats.shp')
patrol_beats['name'] = patrol_beats['name'].apply(str)

gdf_edges['patrol_beat'] = gdf_edges.progress_apply(get_beat_for_edge, patrol_beats = patrol_beats, axis=1)  
```

#### Check that no edges have been left behind

```python
edges_without_beat = gdf_edges[gdf_edges.patrol_beat.isnull()]
print('{} edges belong to no patrol beat'.format(len(edges_without_beat)))

# If some edges were left behind, plot them to see where they are.
if len(edges_without_beat) :
    print(edges_without_beat)
    
    # Plot the graph and polygon
    f, ax = plt.subplots(1, figsize=(30, 30))

    gdf_edges.plot(ax=ax, alpha=1, zorder = 1, legend=True, edgecolor='grey', linewidth=0.2)

    edges_without_beat.plot(alpha=1, ax=ax, linewidth=3,edgecolor='red')

    plt.show()
```

### 2. Nodes

In the case study of Detroit, we don't have incidents at the street level. Instead we have a spatial point that we converted to the closest node in the graph.

In order to calculate `density_hist_inc` for each edge, we need to have an estimation of the number of incidents on each edge.

To do this, we distribute the incidents on each node to the edges that surrounds it, based on their proportion of the total length of the surrounding edges.

We first need to calculate the total length of the surrounding edges (`total_length_adj_edges` in meters) and add this to each node in the gdf.

```python
# get the total length of the surrounding edges
def get_sum_length_edges(row) :
    """ This function calculates and return the sum of the lengths of all edges that either start or end at the node 
    (that is essential for one way roads to be included!)
    There will be duplicated osmid (i.e. all the two way roads u->v v->u) but that is ok. 
    We want to attribute a risk to both those edges in a 2-way road scenario so when agents drive on them in any direction, 
    it measure the deterrence score! Both side of the road will have the same prob_num_inc, as we assume they have the same crime rate
    
    """
    ## Get all edges that either start or end at the node and sum their length attribute
    total_length = gdf_edges[(gdf_edges.u == row.osmid) | (gdf_edges.v == row.osmid)]['length'].sum()
    return total_length

gdf_nodes['total_length_adj_edges'] = gdf_nodes.apply(get_sum_length_edges, axis=1)  
```

# Step 2: Pre-process the patrol beats 

Now that we have created and pre-processed the graph, we need to add a centroid_node to each patrol beat.

### Load the patrol_beats

```python
patrol_beats = gpd.read_file('../data/patrol_beats/patrol_beats.shp')
```

```python
patrol_beats.head()
```

### Get the node closest to the centroid for each beat

```python
patrol_beats.crs
```

```python
centroidSeries = patrol_beats.to_crs('epsg:3857').centroid.to_crs('epsg:4326')
centroidSeries
```

```python
## Add centroid node attribute to patrol_beats

# Find nearest node to cendroid and add column to patrol_beats geodataframe
# Get the nearest node corresponding to the centroid of each patrol beat

def get_nearest_node_centroid(centroid, G):
   
    lon = centroid.x
    lat = centroid.y
    node = ox.get_nearest_node(G, (lat, lon))
    return node

patrol_beats['centroid_node'] = centroidSeries.apply(get_nearest_node_centroid, G = G)
```

# Step 3: preprocess the CFS incidents and crimes

Now that we have created and pre-processed the graph, we need to pre-process our crime and CFS incident datasets to convert the spatial location into a node OSMID. Doing this now instead of in the initialisation step of the ABM environment (`Env`) saves considerable computational time.

The spatial calculation is quite computationally expensive so, due to the size of the datasets, it is necessary to run this script using multiprocessing.


⚠️ You need to run the `dpd_case_study/preprocess_datasets.py` script from a terminal (uses multiprocessing). 
It takes 2 to 4 hours on the Uni of Leeds's HPC (40 nodes).

#### Example

❗️BELOW IS AN EXAMPLE OF HOW THE SCRIPT WORKS ON A SUBSET OF 10 CRIMES

```python
def find_nearest_node(inc, patrol_beats, G ):
    """ Function finds and return the nearest node WITHIN PATROL BEAT to an incident on the graph"""
    #1. Find the patrol beat for the incident
    beat_num = inc.Patrol_beat
    # get the geometry of the patrol beat
    beat_poly = patrol_beats[patrol_beats['name'] == beat_num].geometry.iloc[0]
    
    # truncate G by polygon
    G_beat = ox.truncate.truncate_graph_polygon(G, beat_poly, truncate_by_edge=False)

    node = ox.get_nearest_node(G_beat, (inc.lat, inc.lon))
    return node
     
```

```python
# Import crimes
crimes = pd.read_csv("../data/crimes.csv")[0:10]
crimes['Patrol_beat'] = crimes['Patrol_beat'].apply(str)

# Import patrol_beat
patrol_beats = gpd.read_file('../data/patrol_beats/patrol_beats.shp')

# Apply the find_nearest_node function to each crime
crimes['Node'] = crimes.progress_apply(find_nearest_node, patrol_beats = patrol_beats, G=G, axis=1)  
```
