# ABM of Detroit - node jumping

In this basic ABM implementation, police vehicles move along the road retwork to respond to calls for service. They head back to the station when the incident is sorted.

The movement of the agent is basic: they move from node to node along a route (shortest path to thyeir target). This can be improved later by making them move along the geometry of the graph.

Required libraries:
- OSMNX
- MESA (although can be done without, just need a little tweak)
- Imageio for creating GIFs (although can be done differently, I just haven't worked on it yet)

Model files:
- AgentFramework2: contains class object Agent
- IncidentFramework: contains class object Incident
- ModelFrameword: contains class object Model (main model)

## Alter edge attributes

```python
gdf_nodes, gdf_edges = ox.graph_to_gdfs(G)
print(gdf_nodes.columns)
print(gdf_edges.columns)
```

### Convert maxspeed to kph

```python
# Replace NaN with 25 mph default
gdf_edges.loc[gdf_edges['maxspeed'].isnull(),'maxspeed'] = '25 mph'
gdf_edges['maxspeed'].head()
```

```python
def mph2kmph(mph):
    if type(mph) is list:
        # select first element of the list
        mph = mph[0]
        # extract speed numerical value 
        mph = float(mph.split(" ", 1)[0])
        # convert to kph
        kph = mph * 1.60934
    else:
        # extract speed numerical value 
        mph = float(mph.split(" ", 1)[0])
        # convert to kph
        kph = mph * 1.60934
    return kph
```

```python
%%time
gdf_edges['maxspeed_kph'] = [mph2kmph(mph) for index, mph in gdf_edges['maxspeed'].iteritems()]
```

### Calculate edge driving time (in minutes)

```python
def calculateTime(row) :
    
    # Convert kph to meter per minute
    meters_per_minute = row['maxspeed_kph'] * 1000 / 60 #km per hour to m per minute

    # add an edge attribute for time in minutes required to traverse each edge
    return row['length'] / meters_per_minute
```

```python
%%time
gdf_edges['drive_time'] = [calculateTime(edge) for index, edge in gdf_edges.iterrows()]
gdf_edges['drive_time'].head()
```

## Save all steps as a GIF (for animation)

This takes a while to run if many iterations

```python
def visualise(G, agents_pos, inc_pos, targets_pos, stations_pos) :

    nc = ['g' if node in stations_pos else
          'b' if node in agents_pos else 
          'r' if node in targets_pos else
          '#FFD700' if node in inc_pos else
          'grey' for node in G.nodes()]
    ns = [15 if node in agents_pos or
          node in inc_pos or
          node in targets_pos or
          node in stations_pos else
          1 for node in G.nodes()]
    
    fig, ax = ox.plot_graph(model.graph.G, node_color=nc, node_zorder=3,node_size=ns, fig_height = 10,
                           fig_width = 10, edge_linewidth = 0.3)
    image = fig2array (fig)
    plt.close()
    
    f.value += 1
    
    return image


def fig2array(fig):
    """
    @brief Convert a Matplotlib figure to a 3D numpy array with RGB channels and return it
    @param fig a matplotlib figure
    @return a numpy 3D array of RGB values
    """
    # draw the renderer
    fig.canvas.draw ()
    
    # Get the RGB buffer from the figure
    w,h = fig.canvas.get_width_height()
    buf = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)
    
    # Reshape
    buf.shape = (h, w,3)

    return buf
```

```python
f= widgets.IntProgress(
    value=0,
    min=0,
    max=num_iter, 
    description='Loading:',
)

display(f)
```
